```python
import tkinter as tk  # Tkinter library import garne, GUI banauna
import random  # Random library import garne, dice ko value random garna

# Define snake and ladder positions  # Snake ra ladder ko position haru define garne, game ko rule ma yo important garcha
snakes = {16: 6, 47: 26, 49: 11, 56: 53, 62: 19, 64: 60, 87: 24, 93: 73, 95: 75, 98: 78}
ladders = {1: 38, 4: 14, 9: 31, 21: 42, 28: 84, 36: 44, 51: 67, 71: 91, 80: 100}

# Player positions  # Player haru ko position banauna
player1_pos = 0  # Player 1 ko position 0 ma set huncha, game suru garne
player2_pos = 0  # Player 2 ko position 0 ma set huncha
current_player = 1  # Current player 1 huncha, paila player 1 ko turn huncha

# Roll the dice  # Dice roll garne function
def roll_dice():
    return random.randint(1, 6)  # 1 dekhi 6 samma ko random number return garne, dice ko result random huncha

# Move player based on dice roll  # Dice roll anusar player lai move garne
def move_player(player_pos, dice_value):
    player_pos += dice_value  # Player ko position ma dice ko value jod huncha, normal movement huncha
    if player_pos in snakes:  # Player ko position snake ma chha bhane
        player_pos = snakes[player_pos]  # Snake ko tail ma jancha, game ko rule huncha
    elif player_pos in ladders:  # Player ko position ladder ma chha bhane
        player_pos = ladders[player_pos]  # Ladder ko top ma jancha, game ma udaan huncha
    if player_pos > 100:  # Player ko position 100 bhanda thulo bhayo bhane
        player_pos = 100  # 100 ma matra stop huncha, game ma yo limit huncha
    return player_pos  # Update bhayeko position return garne, player ko position update huncha

# Update the game state when dice is rolled  # Dice roll bhayepachi game ko state update garne
def play_turn(player_number):
    global player1_pos, player2_pos, current_player  # Global variable haru lai use garne
    animate_dice_roll(player_number)  # Dice roll animation call garne

def animate_dice_roll(player_number):  # Dice roll animation function
    for _ in range(10):  # 10 dafa random number display garne
        dice_value = random.randint(1, 6)  # Random dice value generate garne
        dice_label.config(text=str(dice_value))  # Dice label ma current value display garne
        root.update()  # GUI update garne
        root.after(50)  # 50 milliseconds wait garne for animation effect

    # After animation, roll the dice to get the final value
    final_value = roll_dice()  # Final dice value generate garne
    dice_label.config(text=str(final_value))  # Final value display garne
    root.after(500, lambda: update_game_state(player_number, final_value))  # Update game state after a short delay

def update_game_state(player_number, final_value):  # Update game state after rolling
    global player1_pos, player2_pos  # Global player position variables use garne
    result_label.config(text=f"Player {player_number} rolled : {final_value}")  # Result label ma display garne

    # Update player positions based on dice roll
    if player_number == 1:  # If current player 1 ko turn huncha bhane
        player1_pos = move_player(player1_pos, final_value)  # Player 1 lai move garne
        player1_pos_label.config(text=f"Player 1 position: {player1_pos}")  # Player 1 ko position display garne
        if player1_pos == 100:  # If Player 1 le 100 ma pugincha bhane
            result_label.config(text="Player 1 wins!")  # Player 1 ko win ko message display garne
            player1_roll_button.config(state=tk.DISABLED)  # Player 1 ko roll button disable garne, game khatam huncha
    else:  # If current player 2 ko turn huncha bhane
        player2_pos = move_player(player2_pos, final_value)  # Player 2 lai move garne
        player2_pos_label.config(text=f"Player 2 position: {player2_pos}")  # Player 2 ko position display garne
        if player2_pos == 100:  # If Player 2 le 100 ma pugincha bhane
            result_label.config(text="Player 2 wins!")  # Player 2 ko win ko message display garne
            player2_roll_button.config(state=tk.DISABLED)  # Player 2 ko roll button disable garne, game sakkincha

    current_player = 2 if player_number == 1 else 1  # Current player ko turn change garne
    turn_label.config(text=f"Player {current_player}'s turn")  # Current player ko turn display garne

# Reset the game to the initial state 
def reset_game():
    global player1_pos, player2_pos, current_player  # Global variable haru lai use garne
    player1_pos = 0  # Player 1 ko position 0 ma reset huncha, nai game suru garne
    player2_pos = 0  # Player 2 ko position 0 ma reset huncha
    current_player = 1  # Current player 1 ma reset huncha, suru ma player 1 ko turn huncha
    player1_pos_label.config(text="Player 1 position: 0")  # Player 1 ko label update garne
    player2_pos_label.config(text="Player 2 position: 0")  # Player 2 ko label lai grid ma set garne
    result_label.config(text="")  # Result label lai khali huncha, nai game suru garne
    turn_label.config(text="Player 1's turn")  # Turn label update garne, game suru huncha
    player1_roll_button.config(state=tk.NORMAL)  # Player 1 ko roll button lai normal state ma set garne
    player2_roll_button.config(state=tk.NORMAL)  # Player 2 ko roll button lai normal state ma set garne

# GUI window set up garne
root = tk.Tk()  # Tkinter ko main window banauna, game ko interface banauna
root.title("Snake and Ladder Game")  # Window ko title set garne, player lai game ko naam thaha pauna parcha
root.configure(bg='lightblue')  # Window ko background color light blue ma set garne

# Player haru ko position display garne labels
player1_pos_label = tk.Label(root, text="Player 1 position: 0", bg='lightblue', font=("Arial", 14))  # Player 1 ko label
player1_pos_label.grid(row=0, column=0, padx=10, pady=10)  # Label lai grid ma left side ma set garne

player2_pos_label = tk.Label(root, text="Player 2 position: 0", bg='lightblue', font=("Arial", 14))  # Player 2 ko label
player2_pos_label.grid(row=0, column=1, padx=10, pady=10)  # Player 2 ko label lai grid ma right side ma set garne

# Kun player ko turn ho bhanne label
turn_label = tk.Label(root, text="Player 1's turn", bg='lightblue', font=("Arial", 14))  # Initial turn display garne
turn_label.grid(row=1, column=0, columnspan=2)  # Turn label lai grid ma set garne

# Roll buttons for both players  
player1_roll_button = tk.Button(root, text="Player 1 Roll Dice", command=lambda: play_turn(1), bg='yellow', font=("Arial", 12))  # Player 1 ko roll button
player1_roll_button.grid(row=2, column=0, padx=10, pady=10)  # Player 1 ko roll button lai grid ma set garne

player2_roll_button = tk.Button(root, text="Player 2 Roll Dice", command=lambda: play_turn(2), bg='green', font=("Arial", 12))  # Player 2 ko roll button
player2_roll_button.grid(row=2, column=1, padx=10, pady=10)  # Player 2 ko roll button lai grid ma set garne

# Dice ko result display garne label
dice_label = tk.Label(root, text="Roll the Dice", bg='lightblue', font=("Arial", 20))  # Dice ko result display garne
dice_label.grid(row=3, column=0, columnspan=2)  # Dice label lai grid ma set garne

 # Result ko display garne label
result_label = tk.Label(root, text="", bg='lightblue', font=("Arial", 16))  # Result ko display garne label
result_label.grid(row=4, column=0, columnspan=2)  # Result label lai grid ma set garne

# Naya game suru garna ko lagi reset button
reset_button = tk.Button(root, text="Reset Game", command=reset_game, bg='red', font=("Arial", 14))  # Reset button for new game
reset_button.grid(row=5, column=0, columnspan=2)  # Reset button lai grid ma set garne

# Run the GUI loop
root.mainloop()  
```
